```python
from PIL import Image
import sys
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
img = Image.open("berries.jpg")
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
def color_noise(img):
    for i in range(img.width):
        for j in range(img.height):
            r, g, b = img.getpixel((i, j))
            r += random.randint(-150, 150)
            g += random.randint(-150, 150)
            b += random.randint(-150, 150)
            img.putpixel((i, j), (r, g, b))
```

```python
color_noise(img)
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
def white_noise(img):
    for i in range(img.width):
        for j in range(img.height):
            r, g, b = img.getpixel((i, j))
            r += random.randint(-150, 150)
            g += random.randint(-150, 150)
            b += random.randint(-150, 150)
            img.putpixel((i, j), (r, g, b))
```

```python
white_noise(img)
plt.axis("off")
plt.imshow(img)
plt.show()
```
